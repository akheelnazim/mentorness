# In-Depth Analysis

##Run Accumulation by Player:

```python
player_stats = data.groupby('batsman1_name').agg({
    'batsman1_runs': ['sum', 'mean', 'median', 'max', 'count']
}).sort_values(by=('batsman1_runs', 'sum'), ascending=False)
print(player_stats.head())
```

## Wicket Analysis:

```python
# Group data by bowler and match, summing wickets to get total wickets per match
bowler_match_wickets = data.groupby(['bowler1_name', 'match_id']).agg(
    match_wickets=('bowler1_wkts', 'sum')
).reset_index()

# Calculate overall bowler stats
bowler_stats = bowler_match_wickets.groupby('bowler1_name').agg(
    total_wickets=('match_wickets', 'sum'),
    average_wickets_per_match=('match_wickets', 'mean'),
    max_wickets_in_match=('match_wickets', 'max')
).sort_values(by='total_wickets', ascending=False)

# Print the corrected bowler stats
print(bowler_stats.head())
```

## Over-by-Over Progress:

```python
# run rate and wicket fall over the overs to observe patterns and strategies.
sns.lineplot(x='over', y='runs', data=data, estimator='mean', errorbar=None)
plt.title('Average Runs per Over')
plt.xlabel('Over')
plt.ylabel('Runs')
plt.show()
```

# Statistical Insights: Team Strategies and Outcomes

Team Batting and Bowling Strategies:

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Standardize data
features = ['batsman1_runs', 'bowler1_runs', 'bowler1_wkts']
X = data[features].fillna(0)
X_scaled = StandardScaler().fit_transform(X)

# Apply PCA
pca = PCA(n_components=2)
principal_components = pca.fit_transform(X_scaled)
principal_df = pd.DataFrame(data=principal_components, columns=['principal component 1', 'principal component 2'])

sns.scatterplot(x='principal component 1', y='principal component 2', data=principal_df)
plt.title('PCA of Team Strategies')
plt.show()
```
